```python
import gc
import torch
from ultralytics import YOLO
```

```python
model = "yolo26x.pt"

YOLO().train(
        model=f"{model}",
        data="/workspace/datasets/data.yaml",
        project="/workspace/models/20260224",
        name=f"{model}_trained",
        batch=8,
        imgsz=640,
        epochs=100,
        workers=16,
    )
```

```python
def garbage_collector():
    torch.cuda.empty_cache()
    torch.cuda.synchronize()
    gc.collect()

garbage_collector()
```
